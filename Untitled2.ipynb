```python
# Activity 3.01: Building an Artificial Neural Network

# 1. Import required libraries
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset, random_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# 2. Load dataset (UCI Default of Credit Card Clients)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00350/default%20of%20credit%20card%20clients.xls"
df = pd.read_excel(url, header=1)

# Explore
print(df.head())
print(df.shape)
```

```python
# Features & target
X = df.drop(columns=["ID", "default payment next month"]).values
y = df["default payment next month"].values

# Scale features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Convert to tensors
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.long)

# Dataset
dataset = TensorDataset(X_tensor, y_tensor)

# 3. Split: 60% train, 20% val, 20% test
train_size = int(0.6 * len(dataset))
val_size = int(0.2 * len(dataset))
test_size = len(dataset) - train_size - val_size

train_set, val_set, test_set = random_split(dataset, [train_size, val_size, test_size])

train_loader = DataLoader(train_set, batch_size=128, shuffle=True)
val_loader = DataLoader(val_set, batch_size=128)
test_loader = DataLoader(test_set, batch_size=128)
```

```python
# 4. Custom NN class
class CreditNet(nn.Module):
    def __init__(self, input_dim, hidden1=64, hidden2=32, hidden3=16, output_dim=2):
        super(CreditNet, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden1)
        self.fc2 = nn.Linear(hidden1, hidden2)
        self.fc3 = nn.Linear(hidden2, hidden3)
        self.fc4 = nn.Linear(hidden3, output_dim)
        self.relu = nn.ReLU()
        self.logsoftmax = nn.LogSoftmax(dim=1)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.relu(self.fc3(x))
        x = self.logsoftmax(self.fc4(x))
        return x

model = CreditNet(input_dim=X.shape[1])
```

```python
# 5. Train for 50 epochs
criterion = nn.NLLLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

train_losses, val_losses, train_accs, val_accs = [], [], [], []

for epoch in range(50):
    # Training
    model.train()
    total_loss, correct, total = 0, 0, 0
    for xb, yb in train_loader:
        optimizer.zero_grad()
        preds = model(xb)
        loss = criterion(preds, yb)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        correct += (preds.argmax(1) == yb).sum().item()
        total += yb.size(0)
    train_losses.append(total_loss/len(train_loader))
    train_accs.append(correct/total)

    # Validation
    model.eval()
    val_loss, val_correct, val_total = 0, 0, 0
    with torch.no_grad():
        for xb, yb in val_loader:
            preds = model(xb)
            loss = criterion(preds, yb)
            val_loss += loss.item()
            val_correct += (preds.argmax(1) == yb).sum().item()
            val_total += yb.size(0)
    val_losses.append(val_loss/len(val_loader))
    val_accs.append(val_correct/val_total)

    print(f"Epoch {epoch+1}: Train Acc={train_accs[-1]:.3f}, Val Acc={val_accs[-1]:.3f}")
```

```python
# 7. Plot curves
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Val Loss")
plt.legend(); plt.title("Loss Curve")

plt.subplot(1,2,2)
plt.plot(train_accs, label="Train Acc")
plt.plot(val_accs, label="Val Acc")
plt.legend(); plt.title("Accuracy Curve")
plt.show()
```

```python
# Training & validation error rates
train_error = 1 - train_accs[-1]
val_error = 1 - val_accs[-1]
bayes_error = 0.15

high_bias = train_error - bayes_error
high_variance = val_error - train_error

print(f"Train Error: {train_error:.3f}, Val Error: {val_error:.3f}")
print(f"High Bias: {high_bias:.3f}, High Variance: {high_variance:.3f}")
```

```python
# 1. Save model
torch.save(model.state_dict(), "credit_model.pth")

# 2. Load model
loaded_model = CreditNet(input_dim=X.shape[1])
loaded_model.load_state_dict(torch.load("credit_model.pth"))
loaded_model.eval()

# 3. TorchScript conversion
scripted_model = torch.jit.trace(loaded_model, torch.rand(1, X.shape[1]))
scripted_model.save("credit_model_scripted.pt")
```
